# Title

```python
# imports
import skimage as ski
from skimage import restoration
import matplotlib.pyplot as plt
import numpy as np
from im_func import show_image, show_multi_image
import seaborn as sns
from skimage import transform
```

```python
# Load stock image using function
# title = 'camera'
title = 'chelsea'
image = getattr(ski.data, title)()

show_image(image, title)
```

# Add noise

```python
image_noised = ski.util.random_noise(image)
show_image(image_noised, title)
```

# Remove noise

There are several algorithms to remove noise, e.g.:
- total variation (TV)
- bilateral (i.e., smoothing)
- wavelet denoising
- non-local means denoising

## Total variation denoising

```python
image_denoised = ski.restoration.denoise_tv_chambolle(image_noised, weight=0.1, multichannel=True)
fig, ax = plt.subplots(1,3,figsize=[20,6])
plt.sca(ax[0])
show_image(image,'original')
plt.sca(ax[1])
show_image(image_noised,'noised')
plt.sca(ax[2])
show_image(image_denoised, 'denoised')
```

## Bilateral denoising

```python
image_denoised = ski.restoration.denoise_bilateral(image_noised, multichannel=True)
im_title = ((image, 'original'),
            (image_noised, 'noised'),
            (image_denoised, 'denoised'))
show_multi_image(im_title,figsize=[20,8])
```
