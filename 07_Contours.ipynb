# Contours

To generate a list of contours (not just the image of countours). One needs to follow these steps:
- If the image is colored change to gray scale
- threshold the gray scale image
- call the find_contour function (I guess it's wrapper around plt.contour)
- find_contour returns a list of arrays containing the vertices of the individual contours

```python
# imports
import skimage as ski
import skimage.color
import matplotlib.pyplot as plt
import numpy as np
from im_func import show_image, show_multi_image
import seaborn as sns
from skimage import transform
```

```python
# Load stock image using function
title = 'camera'
image = getattr(ski.data, title)()

show_image(image, title)
```

# Convert to gray scale

```python
image_gray = ski.color.rgb2gray(image)
if np.max(image_gray)>1.0:
    image_gray = image_gray/255
show_image(image_gray)
```

# Threshold

```python
from skimage import filters
thr = ski.filters.threshold_isodata(image_gray)
thr = 0.2
image_thr = image_gray>thr
show_image(image_thr)
```

# Contour

```python
contours = skimage.measure.find_contours(image_thr, level=0.5)
show_image(image_thr)
for cont in contours:
    plt.plot(cont[:,1],cont[:,0],'-r')
```
