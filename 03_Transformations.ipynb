```python
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
@interact
def show_resized_image(res=widgets.IntSlider(min=1, max=20, step=1, value=0)):
    plt.imshow(ski.transform.rescale(image,1/res,multichannel=True))
    plt.axis('off')
```

```python
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
from IPython.display import display, clear_output
fig, ax = plt.subplots()

def show_resized_image(res=1.0):
    ax.cla()
    ax.imshow(ski.transform.rescale(image,1/res,multichannel=True))
    ax.axis('off')
    clear_output(wait = True)
    display(fig)
    
for i in range(30):
    _ = show_resized_image(res=i+1)
```
