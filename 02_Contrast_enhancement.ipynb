# Contrast enhancement

Several techniques can be used:
- contrast stretching: stretch the histogram so the full range of intensity values is filled
- histogram equalization: spreads out the most fequent intensity values using probability distribution

Histogram equalization has three types:
- standard
- adaptive
contrast limited adaptive histogram equalizatn (CLAHE)

```python
# imports
import skimage as ski
import matplotlib.pyplot as plt
import numpy as np
from im_func import show_image
import seaborn as sns
```

```python
# Load stock image using function
title = 'camera'
title = 'coffee'
image = getattr(ski.data, title)()

show_image(image, title)
```

```python
# Plot the istogram
hist = ski.exposure.histogram(image)
fig, ax = plt.subplots(1,2,figsize=[10,5])
plt.sca(ax[0])
show_image(image, title)
plt.sca(ax[1])
plt.plot(hist[1],hist[0])
```

```python
# Standard histogram equalization
im_eq = ski.exposure.equalize_hist(image)
fig, ax = plt.subplots(1,2,figsize=[10,5])
plt.sca(ax[0])
show_image(image, 'original')
plt.sca(ax[1])
show_image(im_eq, 'equalized')
fig, ax = plt.subplots(1,1,figsize=[10,5])

hist = ski.exposure.histogram(image)
hist_eq = ski.exposure.histogram(im_eq)
plt.plot(hist[1],hist[0],'b',label='original')
plt.plot(hist_eq[1]*255,hist_eq[0],'r',label='equalized')
```

```python
# CLAHE histogram equalization
im_eq = ski.exposure.equalize_adapthist(image,clip_limit=0.01)
fig, ax = plt.subplots(1,2,figsize=[10,5])
plt.sca(ax[0])
show_image(image, 'original')
plt.sca(ax[1])
show_image(im_eq, 'equalized')
fig, ax = plt.subplots(1,1,figsize=[10,5])

hist = ski.exposure.histogram(image)
hist_eq = ski.exposure.histogram(im_eq)
plt.plot(hist[1],hist[0],'b',label='original')
plt.plot(hist_eq[1]*255,hist_eq[0],'r',label='equalized')
```
